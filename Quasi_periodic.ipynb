```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy import signal
import sys
!{sys.executable} -m pip install scikit-dsp-comm
import sk_dsp_comm.sigsys as ss
import sk_dsp_comm.digitalcom as dc
```

```python
t1 = np.arange(0,8,.01)
x1 = signal.square(2*np.pi*t1)
t2 = np.arange(8,10,.01)
x2 = 0.2*t2 -1
t3 = np.arange(10,18,.01)
x3 = signal.square(2*np.pi*t3)

t = np.concatenate([t1,t2,t3])
x = np.concatenate([x1,x2,x3]) + np.random.normal(0,0.2,len(t))

plt.plot(t,x)
plt.grid()
```

```python
Ryx,tau = dc.xcorr(x,x,1000) #note order change
# We know Ryx is real
Ryx = Ryx.real
plt.plot(tau,Ryx)
plt.title(r'Auto Correlation')
plt.xlabel(r'Autocorrelation Lag $\tau$ (s)')
plt.grid();
```
